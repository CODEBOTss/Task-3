```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales': [500, 600, 800, 700, 900],
    'Profit': [50, 60, 80, 70, 90]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)  
plt.bar(df['Year'], df['Sales'], color='blue', label='Sales')
plt.xlabel('Year')
plt.ylabel('Sales ($)')
plt.title('Yearly Sales')
plt.legend()

plt.subplot(2, 1, 2)  
plt.plot(df['Year'], df['Profit'], marker='o', color='green', linestyle='-', label='Profit')
plt.xlabel('Year')
plt.ylabel('Profit ($)')
plt.title('Yearly Profit')
plt.legend()

plt.tight_layout() 
plt.show()
```
